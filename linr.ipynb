```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression 
from sklearn.metrics import r2_score

data_train = pd.read_csv('test.csv')

x_train = data_train['x']
y_train = data_train['y']
A = []
B = []
for x in x_train:
    A.append([1,x])
X = np.array(A)
for y in y_train:
    B.append(y)
Y = np.array(B)
def lin(X,Y):
    X_trans = np.transpose(X)
    #print(X_trans[1])
    X_final = np.dot(X_trans,X)
   
    X_final = np.linalg.inv(X_final)
    
    X_final = np.dot(X_final,X_trans)
    
    wlin = np.dot(X_final, Y)
    return wlin
w = lin(X,Y)

y_plot = []
for i in range(100):
    y_plot.append(w[0] + w[1] * i)
plt.figure(figsize=(10,10))
plt.scatter(x_train,y_train,color='red',label='GT')
plt.plot(range(len(y_plot)),y_plot,color='black',label = 'pred')
plt.legend()
plt.show()
```
